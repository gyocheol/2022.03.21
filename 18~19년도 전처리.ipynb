```python
e_18 = pd.read_excel('datasets/전력사용량/(2018)시군구별 전력사용량.xlsx', header=2)
e_19 = pd.read_excel('datasets/전력사용량/(2019)시군구별 전력사용량.xlsx', header=2)
e_18
```

```python
# 서울 부산 대구 인천 광주 대전 울산 : 광역시별로 결정!
condition = (e_18.시도 == '서울특별시')|(e_18.시도 =='부산광역시')|(e_18.시도 == '대구광역시')|(e_18.시도 == '인천광역시')|(e_18.시도 == '광주광역시')|(e_18.시도 == '대전광역시')|(e_18.시도 == '울산광역시')# 조건식 작성

e_18 = e_18[condition]
e_18
```
